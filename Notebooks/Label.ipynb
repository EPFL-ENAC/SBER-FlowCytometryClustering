```python
#Summary

#Files
file = '/GL53_UP_2A.fcs'
folder_all = '/../data_gated/all_event'
folder_gated = '/../data_gated/gated'

#Load to DataFrame
fc_data_all, channels = load_important_data(file,folder_all)
fc_data_all = np.array(fc_data_all)
df_all = pd.DataFrame(data=fc_data_all, columns=['FSC-H', 'SSC-H', 'B530-H', 'B572-H', 'B675-H'])

fc_data_gated, channels = load_important_data(file,folder_gated)
fc_data_gated = np.array(fc_data_gated)
df_gated = pd.DataFrame(data=fc_data_gated, columns=['FSC-H', 'SSC-H', 'B530-H', 'B572-H', 'B675-H'])

#create a hash for each entry to find all entries not gated
df_all.loc[:, "hash"] = df_all.apply(lambda x: hash(tuple(x)), axis = 1)
df_gated.loc[:, "hash"] = df_gated.apply(lambda x: hash(tuple(x)), axis = 1)
df_not_gated = df_all.loc[~df_all["hash"].isin(df_gated["hash"]), :]

#label dataset
df_gated.loc[:,"label"] = 3
df_not_gated.loc[:,"label"] = 1

#concat dataset and remove hash column
#df_final = pd.concat([df_gated, df_not_gated], ignore_index=True)
df_final = df_not_gated.append(df_gated,ignore_index=True)
df_final.drop('hash', inplace=True, axis=1)

#creation of X and y
X = df_final.drop('label',axis=1)
y = df_final['label']
```
